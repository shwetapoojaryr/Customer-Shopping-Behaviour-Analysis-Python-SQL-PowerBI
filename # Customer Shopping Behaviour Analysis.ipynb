# Customer Shopping Behaviour Analysis

```python
import pandas as pd
```

```python
df =  pd.read_csv('customer_shopping_behavior.csv')

df.head()
```

```python
df.info()
```

```python
df.describe(include = 'all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')[ 'Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# for better readability and analysis change to snake case
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_') 
df = df.rename(columns = {'purchase_amount_(usd)' : 'purchase_amount'})
```

```python
df.columns
```

```python
# create a column age_group

labels = ['Young Adult','Adult','Middle-Aged','Senior']
df['age_group'] = pd.qcut(df['age'], q = 4, labels = labels) #splits ages into four equal sized groups and assign labels defined 
```

```python
df[['age','age_group']].head(10)
```

```python
# create column purchase_frequency_days

frequncy_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequncy_mapping) # replace the text to corresponding no of in columns
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all() #is any one of the column redundant or not
```
